## **NLP** (Natural Lenguage Processing)

NLP braches:
- NLU (Natural Language Understanding) . for content and context (sentiment analysis, summarization)
- Information extraction to extract structured information from text (entities,search)
- Natural Langage Generation, to generate human conversation like text
- ASR (Automatic Speech Recognition) to convert voice into text

NLP training steps
- train on text corpus {'text': 'This course is nice'}
- label and annotation {'label': 'POSITIVE', 'text': 'This course is nice'}
- Tokenization convert words, symbols or characters {'tokens': ['This', 'course', 'is', 'nice']}
- Vectorization {'label':1, 'vocab_ids':{234,123,567,112}}
- Build the model {architecture, prameters, metrics}

Inference for NLP
- input text {'text': 'This sandwich tastes great'}
- Tokenization convert words, symbols or characters {'tokens': ['This', 'sandwich', 'tastes', 'great']}
- Vectorization {'vocab_ids':{234,143,899,543}}
- Predict label logits [0.95, 0.05] delivery probabilities
- Decode label predicted label: "POSITIVE"

```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
classifier('This is a great day')
```

```python
result = classifier('This is a bad day')
result[0]['label']
```

```python
text_generator = pipeline("text-generation")
result = text_generator("This is a great day for ", max_length=20, truncation=True, do_sample=False)
result
```

```python
qu_an = pipeline("question-answering")
context = """
Natural language processing (NLP) is a subfield of linguistics, computer science,
and artificial intelligence concerned with the interactions between computers and
human language, in particular how to program computers to process and analyze
large amounts of natural language data. The result is a computer capable of
"understanding" the contents of documents, including the contextual nuances of
the language within them. The technology can then accurately extract information
and insights contained in the documents as well as categorize and organize the
documents themselves.
"""
result = qu_an(question="What is NLP?", context=context)
result
```
